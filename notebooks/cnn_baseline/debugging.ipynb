```python
import os
spectrogram_folder = os.path.abspath("../../data/processed/spectrograms/xeno_canto/")
spectrogram_files = os.listdir(spectrogram_folder)
print("Example spectrogram filenames:", spectrogram_files[:10])
print(f"Total files in spectrogram folder: {len(spectrogram_files)}")
```

```python
example_id = "20356"  # Replace with an ID from your parquet file
matches = [f for f in spectrogram_files if f.startswith(example_id)]
print(f"Files matching ID {example_id}: {matches}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_spectrogram(file_path):
    S_db = np.load(file_path)
    plt.figure(figsize=(10, 4))
    plt.imshow(S_db, aspect='auto', origin='lower', cmap='magma')
    plt.tight_layout()
    plt.axis('off')
    plt.show()

spectrogram_path = os.path.abspath("../../data/processed/spectrograms/xeno_canto/585887_logmel.npy")
visualize_spectrogram(spectrogram_path)
```
