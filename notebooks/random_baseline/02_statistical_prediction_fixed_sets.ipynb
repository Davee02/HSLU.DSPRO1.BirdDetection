```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

train_file_location = "../../data/cleaned/70_15_15_cleaned_train.csv"
val_file_location = "../../data/cleaned/70_15_15_cleaned_val.csv"
train_df = pd.read_csv(train_file_location)
val_df = pd.read_csv(val_file_location)

y_train = train_df['en']
y_val = val_df['en']

bird_counts = y_train.value_counts(normalize=True)

predicted_classes = y_val.map(bird_counts)

predicted_classes = predicted_classes.fillna(0)

y_val = y_val.astype(str)
predicted_classes = predicted_classes.astype(str)

accuracy = accuracy_score(y_val, predicted_classes)
precision = precision_score(y_val, predicted_classes, average='weighted', zero_division=1)
recall = recall_score(y_val, predicted_classes, average='weighted', zero_division=1)
f1 = f1_score(y_val, predicted_classes, average='weighted', zero_division=1)
conf_matrix = confusion_matrix(y_val, predicted_classes)

print("Baseline Model Performance (Single Train-Validation Split):")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

print("\nClassification Report:")
print(classification_report(y_val, predicted_classes, zero_division=1))

print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

train_file_location = "../../data/cleaned/70_15_15_cleaned_train.csv"
val_file_location = "../../data/cleaned/70_15_15_cleaned_val.csv"
train_df = pd.read_csv(train_file_location)
val_df = pd.read_csv(val_file_location)

y_train = train_df['en']
y_val = val_df['en']

bird_counts = y_train.value_counts(normalize=True)

y_pred_probabilities = y_val.map(bird_counts).fillna(0)

most_probable_bird = bird_counts.idxmax()
predicted_classes = y_pred_probabilities.apply(lambda prob: most_probable_bird if prob == 0 else bird_counts.idxmax())

y_val = y_val.astype(str)
predicted_classes = predicted_classes.astype(str)

accuracy = accuracy_score(y_val, predicted_classes)
precision = precision_score(y_val, predicted_classes, average='weighted', zero_division=1)
recall = recall_score(y_val, predicted_classes, average='weighted', zero_division=1)
f1 = f1_score(y_val, predicted_classes, average='weighted', zero_division=1)
conf_matrix = confusion_matrix(y_val, predicted_classes)

print("Baseline Model Performance (Single Train-Validation Split):")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

print("\nClassification Report:")
print(classification_report(y_val, predicted_classes, zero_division=1))

print("\nConfusion Matrix:")
print(conf_matrix)
```
